<a name="ex1-1" style="color:red">**Exercise 1-1**</a> **[1 point]** What are the total number of word tokens in this corpus? In this case, consider a word to be any string that is separated by whitespace. What are the total number of distinct word _types_?

```python
# open the file
file = open("sec02-21.raw")
```

```python
def split_words_on_regex(string, regex):
    
    # split string on delimiter
    words = re.split(regex, string)
    
    #  last element is always an empty string so remove it
    return words[:len(words)-1]

def pretty_print(words):
    
    # print the number of tokens (every word is one token)
    print("number of word tokens: \t", len(words))
    
    # print the number of unique tokens (remove duplicates)
    print("number of word types: \t", len(set(words)))
```

```python
# read the entire corpus
corpus = file.read() 

# convert the corpus to a single string
string = "".join(corpus)

# split the corpus on whitespaces
words_on_whitespace = split_words_on_regex(string,"\s+")

# print the results
pretty_print(words_on_whitespace)
```

<a name="ex1-4" style="color:red">**Exercise 1-4**</a> **[2 points]** Plot a graph of word frequency versus rank of a word in this corpus (after removing punctuation and lower-casing). Does this corpus obey Zipf's law? 

Yes, as seen below, this corpus obeys Zipf's law. 

```python
import operator

def count_words(words):
    
    # count the words
    count = Counter(words)
    
    # sort the words from highest to lowest (first one is the highest rank)
    sorted_count = sorted(count.items(), key=operator.itemgetter(1), reverse=True)
    
    # only return the word counts, not the words themselves
    return [x[1] for x in sorted_count]
```

```python
def plot_ranks_occurences(words):
    
    # get the word counts
    counts = count_words(words)
    
    # use this subplot to set the scale to a logarithm for a better curve
    ax = plt.subplot()
    ax.set_yscale("log") 
    
    # plot the occurences
    plt.plot(np.arange(len(counts)), counts)
    plt.ylabel('word occurences')
    plt.xlabel('ranked words')
    plt.show()
```

```python
plot_ranks_occurences(words_lowercase)
```

<a name="ex1-5" style="color:red">**Exercise 1-5**</a> **[1 point]** While processing the corpus as above, there may still be a few things that you don't consider real words. What are some of those things? [The question of what counts as a word and what doesn’t comes up a lot in NLP, and there is rarely a single correct answer; often we need to use our judgment and justify our decisions.]. Go back and remove some of these non-words, and re-plot the graph. 

Single letters are usually not words, with an exception of 'a' and 'I'. The other change is in the regex. The numbers, like '21' or '4', are removed. Words like '1920s' stay, since it is a reference.

```python
def highest_ranked_words(words):
    
    # count the words
    count = Counter(words)
    
    # sort the words from highest to lowest (first one is the highest rank)
    sorted_count = sorted(count.items(), key=operator.itemgetter(1), reverse=True)
    
    # only return the words
    return [x[0] for x in sorted_count]

print(highest_ranked_words(words_lowercase))
```

```python
def remove_words(old, remove):
    
    # create a hard copy
    new = old.copy()
    
    # go over every word that needs to be removed
    for word in remove:
        
        # while the word is in the list
        while word in new:
            
            # remove the word
            new.remove(word)
    
    # return the result
    return new
```

```python
# split the corpus on non alphanumeric characters
cleaned_words = remove_words(words_non_numeric, ['s', 'n', 't', 'u', 'nt', 'm'])
```

<a name="ex1-6" style="color:red">**Exercise 1-6**</a> **[0.5 point]** What is the average _sentence length_ of sentences in the Penn Treebank? How does it change along the pre-processing steps (tokenization, lowercasing, other transformations).

The average lengths are found below. The lowercasing has a small to no change to the average sentence length. Only when the "invalid" words are removed. That means that some or these "words" contain uppercases. The tokenization has the biggest influence on the average sentence length, as expected, since a lot of non words are removed.

```python
def average_sentence_length(corpus, regex, clean):
    
    # the list of sentence lenghts
    sentence_lengths = []
    
    # the sentences of the corpus
    sentences = split_words_on_regex(corpus, "\W+\.\W+")
    
    # go over every sentence
    for sentence in sentences:
        
        # split the sentence on the words (on the given regex)
        words = split_words_on_regex(sentence, regex)
        
        # remove the non words if we want to clean the sentences
        if (clean):
            words = remove_words(words, ['s', 'n', 't', 'u', 'nt', 'm'])
        
        
        # store the number of words as the length of the sentence
        sentence_lengths.append(len(words))
    
    # return the average sentence length
    return sum(sentence_lengths)/len(sentence_lengths)
```

```python
def all_senctence_lengths(corpus, regex):
    print("Normal: ", average_sentence_length(corpus, regex, False))
    print("Lowercase: ", average_sentence_length(corpus.lower(), regex, False))
    print("Cleaned: ", average_sentence_length(corpus, regex, True))
    print("Cleaned lowercase: ", average_sentence_length(corpus.lower(), regex, True))
```

```python
# split on whitespaces
all_senctence_lengths(string, "\s+")
```

```python
# split on non alphanumeric
all_senctence_lengths(string, "\W+")
```

```python
# split on non alphanumeric and remove numbers
all_senctence_lengths(string, "\W+\d+\W+|\W+")
```
